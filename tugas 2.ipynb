```python
#Data Transformasi
#digunakan untuk memilih kolom-kolom yang berisi tipe data numerik dari DataFrame 
numeric_columns = df.select_dtypes(include=['number']).columns
df[numeric_columns]
```

```python
#nilai NaN telah diganti dengan nilai rata-rata dari kolom tersebut.
for column in numeric_columns:
    column_mean = df[column].mean()
    df[column].fillna(column_mean, inplace=True)
df
```

```python
#untuk memilih dan mengakses kolom-kolom dalam DataFrame df yang bertipe data object
object_columns = df.select_dtypes(include=['object']).columns
df[object_columns]
```

```python
#mengonversi nilai non-numerik dalam kolom GPA dari DataFrame df menjadi NaN (
df['GPA'] = pd.to_numeric(df['GPA'], errors='coerce')
```

```python
#menghitung nilai rata-rata (mean) dari kolom GPA dalam DataFrame df.
mean_GPA = df['GPA'].mean()
```

```python
# untuk mengganti nilai NaN dalam kolom GPA dengan nilai rata-rata (mean) yang telah dihitung sebelumnya.
df['GPA'].fillna(mean_GPA, inplace=True)
```

```python
#mendapatkan daftar nilai unik (tanpa duplikasi)
df['GPA'].unique()
```

```python
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
```

```python
mean_weight = df['weight'].mean()
```

```python
df['weight'].fillna(mean_weight, inplace=True)
```

```python
#menghitung dan menampilkan jumlah nilai unik yang terdapat dalam setiap kolom bertipe object di DataFrame
object_columns = df.select_dtypes(include=['object']).columns
df[object_columns]
```

```python
# menghitung dan menampilkan jumlah kemunculan setiap nilai dalam kolom bertipe object di DataFrame
for column in object_columns:
    print(f"Value counts for column '{column}':")
    print(df[column].value_counts().sort_values(ascending=False))
    print()
```

```python
#untuk mengonversi nilai kategorikal dalam kolom fav_cuisine menjadi nilai numerik menggunakan pemetaan (mapping) yang telah didefinisikan.
cuisine_mapping_numeric = {
    'Italian': 1,
    'Mexican': 2,
    'American': 3,
}
# Encode the 'fav_cuisine' column with numerical values
df['fav_cuisine_encoded'] = df['fav_cuisine'].map(cuisine_mapping_numeric).fillna(0)
# Display the DataFrame with the new encoded column
df
```

```python
#digunakan untuk mengakses kolom baru dalam DataFrame df yang telah ditambahkan sebelumnya, yaitu kolom fav_cuisine_encoded. 
#Kolom ini berisi nilai numerik yang dihasilkan dari pemetaan (mapping) nilai-nilai dalam kolom fav_cuisine
df['fav_cuisine_encoded']
```

```python
#untuk mengidentifikasi kolom-kolom numerik dalam DataFrame df dan membuat DataFrame baru yang hanya berisi kolom-kolom tersebut.
numeric_columns = []
for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        numeric_columns.append(column)
cleaned_df=df[numeric_columns]
```
